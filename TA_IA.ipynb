```python
import mss
import numpy as np
import cv2
import pygetwindow as gw
from ultralytics import YOLO

# ======================
# CARGAR MODELO YOLO
# ======================
model = YOLO(r"C:\Users\r2bvf\Documents\1IIEE13 - Fabricacion de equipos electronicos (2024-2)\runs\detect\train\weights\best.pt")

# ======================
# TTULO DE LA VENTANA A CAPTURAR
# ======================
window_title_substring = "Revisa"   # Usa parte del ttulo de tu pestaa

# Buscar ventana
windows = gw.getWindowsWithTitle(window_title_substring)
if not windows:
    raise Exception(f"No se encontr ninguna ventana que contenga: {window_title_substring}")

window = windows[0]  # Tomar la primera coincidencia
print(f"Capturando ventana: {window.title}")

# Obtener posicin y tamao de la ventana
region = {
    "top": window.top,
    "left": window.left,
    "width": window.width,
    "height": window.height
}

print("Regin detectada:", region)

# ======================
# LOOP PRINCIPAL
# ======================
with mss.mss() as sct:
    while True:
        # Capturar la ventana
        frame = np.array(sct.grab(region))

        # Convertir BGRA  BGR
        frame = cv2.cvtColor(frame, cv2.COLOR_BGRA2BGR)

        # YOLO deteccin
        results = model(frame)[0]

        # Dibujar detecciones
        for box in results.boxes:
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
            conf = float(box.conf)
            cls = int(box.cls)

            cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)
            cv2.putText(
                frame,
                f"{model.names[cls]} {conf:.2f}",
                (x1, y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.6,
                (0,255,0),
                2
            )

        # Mostrar ventana de deteccin (NO la captura)
        cv2.imshow("YOLO - Screen Capture (window)", frame)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

cv2.destroyAllWindows()
```
